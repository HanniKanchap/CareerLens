```python
rate = 86.14

def exchange(salary):
    if 'competitive' in salary.lower():
        return pd.Series([None, None])

    is_usd = '$' in salary
    salary = salary.replace('$', '').replace('₹', '').replace('/year', '').replace(',', '').strip()
    parts = salary.split('-')
    try:
        min_sal = float(parts[0].strip())
        max_sal = float(parts[1].strip()) if len(parts) > 1 else min_sal

        if is_usd:
            min_sal *= rate
            max_sal *= rate

        return pd.Series([int(min_sal), int(max_sal)])
    except Exception as e:
        print(f"Error parsing salary: {salary} -> {e}")
        return pd.Series([None, None])

df[['Min_Salary', 'Max_Salary']] = df['Salary'].apply(exchange)
```

```python
category_map = {
    'Data & AI': ['data', 'ml', 'machine learning', 'ai', 'artificial intelligence', 'analytics', 'vision','analyst','power bi','tableau','python'],
    'Software Development': ['ios','android','kotlin','developer', 'engineer', 'programmer', 'full stack', 'backend', 'front end','frontend' ,'web development','app development'],
    'Cybersecurity': ['cyber', 'security', 'soc', 'iam'],
    'Quality Assurance & Testing': ['qa', 'quality', 'test', 'sdet'],
    'DevOps & Cloud': ['devops', 'aws', 'azure', 'cloud', 'gcp', 'platform'],
    'Game Development': ['game', 'unity'],
    'UI/UX & Design': ['ux', 'ui', 'figma'],
    'IT Support & Infrastructure': ['support', 'administrator', 'admin', 'it', 'desktop', 'infrastructure'],
    'SAP & ERP': ['sap', 'odoo', 'd365', 'oracle', 'erp'],
    'Sales & Marketing': ['sales', 'marketing', 'seo', 'content', 'digital'],
    'Project & Product Management': ['project manager', 'product manager', 'coordinator', 'scrum','mamager','consultant'],
    'Education & Training': ['trainer', 'instructor', 'teacher', 'faculty'],
    'Electronics & Hardware': ['electronics', 'circuit', 'hardware', 'embedded','iot']
}
def categorize(title):
    title_lower = title.lower()
    for category, keywords in category_map.items():
        if any(keyword in title_lower for keyword in keywords):
            return category
    return 'Other'

df['Category'] = df['Job Title'].apply(categorize)
```

```python
def categorize(title):
    title_lower = title.lower()
    for category, keywords in category_map.items():
        if any(keyword in title_lower for keyword in keywords):
            return category
    return 'Other'

df['Category'] = df['Job Title'].apply(categorize)
```

```python
# Precompute category-wise average salary
category_avg = df[['Category', 'Min_Salary', 'Max_Salary']].assign(Avg_Salary=lambda d: d[['Min_Salary', 'Max_Salary']].mean(axis=1)).groupby('Category')['Avg_Salary'].mean()

def impute_salary(row):
    if pd.isna(row['Min_Salary']) or pd.isna(row['Max_Salary']):
        avg = category_avg.get(row['Category'])
        return pd.Series([avg, avg])
    return pd.Series([row['Min_Salary'], row['Max_Salary']])

df[['Min_Salary', 'Max_Salary']] = df.apply(impute_salary, axis=1)
```

```python
df['Min_Salary'] = df['Min_Salary'].apply(lambda x: round(x, 2)).astype(int)
df['Max_Salary'] = df['Max_Salary'].apply(lambda x: round(x, 2)).astype(int)
```

```python
df = df.reset_index(drop=True)
```
