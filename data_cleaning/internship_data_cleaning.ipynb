```python
internship_categories = {
    "Testing & QA": [
        "software testing", "quality assurance", "qa","quality analyst" ,"manual testing","testing"
        "automation testing", "mobile app testing","management","IT support","system admin"
    ],
    "Software and web Development": [
        "angular","full stack", "ruby", "software development","developer", "software engineer","react","native","Front End",
        "development engineer", "founding", "interim", "junior software","kotlin","development",
        "trainee", "flutter", "mern", "kotlin", "frontend", "backend","php","Product Development","c","c++","django","java", "web", "wordpress", "react", "native", "js", "app","microsoft","laravel","Development Engineer",".net"
    ],
    "AI / ML / Data Science": [
        "ml", "genai", "ai", "llm","bot", "chatbot", "data automation",
        "data scientist", "data analytics", "llvm", "compiler", "research","sql",
        "computer vision", "analyst","data","mathematics"
    ],
    "Cloud, DevOps & Infrastructure": [
        "cloud", "aws", "azure", "devops", "infrastructure"
    ],
    "Game & Graphics Development": [
        "game", "unity", "unreal", "3d", "animator","2D"
    ],
    "Cybersecurity & InfoSec": [
        "cyber security", "penetration testing", "infosec", "security operations",
        "information security"
    ],
    "Embedded Systems & Hardware": [
        "embedded", "electronics", "field engineer", "hardware" , 'iot'
    ],
    "UI/UX & Design": [
        "ui","ux", "user interface", "user experience", "design", "ui design",
        "ui designer", "ux researcher", "visual design", "interface"
    ],
    "Training & Teaching": [
        "trainer", "instructor", "teaching", "tutor", "mentor", "game development instructor","instructor"
    ],
    "Business, Content & Misc": [
        "marketing", "business associate", "proposal", "content","entrepreneur","entrepreneurship",
        "wordpress implementation", "sales", "digital transformation",
        "photo collection", "pnp", "non-tech", "student", "technical support"
    ]
}

def extract_category(title):
    title = str(title).lower().strip()
    for cat, words in internship_categories.items():
        for keyword in words:
            if keyword.lower() in title:
                return cat
    return "Others"

data['Category'] = data['Job Title'].apply(extract_category)
```

```python
data.dropna(inplace= True)
```

```python
data.rename(columns={'location':'Location'},inplace=True)
```

```python
data['Location'] = data['Location'].str.split(',')
data['Location'] = data['Location'].apply(lambda x: [i.strip() for i in x])

data = data.explode('Location')
```
